```python
f = open('/home/jittat/Dropbox/adm64/project-majors/17-sport.csv','r')

def extract_credit(st):
    items = st.split()
    return float(items[-2])

reader = csv.reader(f)
for r in reader:
    if len(r) > 5:
        req = r[5]
        lines = [l for l in  req.split("\n") if (('กลุ่ม' in l) and ('กิต' in l))]
        
        sci_lines = [l for l in lines if 'วิท' in l]
        math_lines = [l for l in lines if 'คณิต' in l]
        lang_lines = [l for l in lines if 'ภาษา' in l]
        #print(lines)
        #print(sci_lines)
        #print(math_lines)
        #print(lang_lines)
        
        if (len(sci_lines) > 1) or (len(math_lines) > 1) or (len(lang_lines) > 1):
            print('ERROR')
            continue
            
        sci_credit = 0
        math_credit = 0
        lang_credit = 0
        if len(sci_lines) == 1:
            sci_credit = extract_credit(sci_lines[0])
        if len(math_lines) == 1:
            math_credit = extract_credit(math_lines[0])
        if len(lang_lines) == 1:
            lang_credit = extract_credit(lang_lines[0])
        
        print(f'{r[0]},{sci_credit}/{math_credit}/{lang_credit}')
        
```
